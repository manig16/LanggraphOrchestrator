```python
# Demo: initialize auth/LLM, load data, build workflow, and process records
from pathlib import Path
import json, re

from auth import init_auth, make_llm_client
from tools import set_llm_client, set_reference_codes
from loader import load_data
from orchestrator import build_workflow, process_claim

init_auth('./Data/claimaudit.env')
llm = make_llm_client()
set_llm_client(llm)

data = load_data(Path('./Data'))
set_reference_codes(data.get('reference_codes', {}))

# Build the LangGraph workflow
app = build_workflow()

# Read test data and run the workflow for each and get the decision
with open('data/test_data.json', 'r') as f:
    records = json.load(f)

for idx, rec in enumerate(records, start=1):
    result = process_claim(rec, app)
    print(f"{idx}: patient={rec.get('patient_id')} decision={result}")
```
